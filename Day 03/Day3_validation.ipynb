```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15, 5]
```

```python
real_data = pd.read_csv('./data-storm-30/validation_data.csv')
predicted_data = pd.read_csv('./out.csv')
```

```python
predicted_data
```

```python
predicted_data['CategoryCode'] = predicted_data["ID"].str[:10]
predicted_data['Week'] = predicted_data["ID"].str[-2:]
predicted_data['ItemCode'] = pd.to_numeric(predicted_data["ID"].str[11:-3])
predicted_data.drop('ID',axis='columns', inplace=True)
```

```python
predicted_data.drop('CategoryCode',axis='columns', inplace=True)
real_data.drop('CategoryCode',axis='columns', inplace=True)
```

```python
gpr = predicted_data.groupby('ItemCode')
gre = real_data.groupby('ItemCode')
```

```python
gpr2 = {i: {w: s for (_, (s, w, _)) in weeks.iterrows()} for (i, weeks) in gpr }
gre2 = {i: {w: s for (_, (_, w, s)) in weeks.iterrows()} for (i, weeks) in gre }
```

```python
gpr2
```

```python
gre2
```

```python
te = 0
ts = 0

for item_code in gre2.keys():
    for w in gre2[item_code].keys():
        te += abs(gre2[item_code][w] - gpr2[item_code][w])
        ts += gre2[item_code][w]
```
